```python
#1 append-means add
list1=[1,3,4,5,6,6]
list2=[1,3,4,5,6,6]
list3=[1,3,4,5,6,6]
list4=[1,3,4,5,6,6]
list6=["a","b","c"]
list5=[1,3,4,5,6,6]
list7=[1,3,4,5,6,6]
list8=[1,3,4,5,6,6]
list9=[1,3,4,5,6,6]
list10=[1,3,4,5,6,6]
list11=[1,3,4,5,6,6]


list1.append(2)
print(list1)

#2 clear
list2.clear()
print(list2)

#3 copy
list3.copy()
print(list3)

#4 count
print(list4.count(1))
print(list4)

#5 extend
list5.extend("a")
print(list5)

#6 index
print(list6.index("a"))
print(list6)

#7 insert
list7.insert(0,2)
print(list7)

#8 pop
list8.pop()
print(list8)

#9 remove
list9.remove(1)
print(list9)

#10 reversal
list10.reverse()
print(list10)

#11 sort
list11.sort()
print(list11)
```
