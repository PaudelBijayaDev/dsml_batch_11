# Tuples Functions in Python

```python
T=(5,'Bijaya', 49.32,[4,8,12])

#len()
print(len(T))

lst=(2,3,4,5)
#max()
print(max(lst))

#min()
print(min(lst))

#sum()
print(sum(lst))

#any()
print(any(T))#gives bool op

#all()
print(all(T))#gives bool op

#sorted()
print(sorted(lst))

#count()
print(T.count(5))

#index()
print(T.index(5))

#tuple()
print(tuple(T))


```

# Packing and unpacking in python

```python
#packing
tup=(4,3,2,'shree')
tup

#unpacking
tup=(2,4,6)
a,b,c = tup

print(a)
print(b)
print(c)
```

```python
#skipping data using dash
data1=(4,5,6)
l,b,_=data1
print(l)
print(b)
```

# Set Operation

```python
A={1,2,3}
B={3,6,7}
print(A|B)#union
print(A&B)#intersection
print(A-B)#difference
print(A^B)#intersection complement
print(2 in A)#membership
print(A<=B)#subset
print(A>=B)#superset
```
