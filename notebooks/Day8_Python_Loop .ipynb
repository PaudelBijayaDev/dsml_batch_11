# while loop

```python
#while expression:statements

#sum of 10 numbers

sum=0
i=1

while i<=10:
    sum=sum+i
    i+=1

print('the sum of 10 numbers is',sum)
```

# break

```python
#break-breaks the loop when condn meets
i=1
while i<=10:
  print('number is')
  i+=1

  if i==3:
    break
  print(i)
```

# continue

```python
#continue-continues the statement loop
i=1
while i<=10:
  print('number is')
  i+=1

  if i==3:
    continue
  print(i)
```

# Pass

```python
#passes the loop or simply pass no interrupt(condition. is not checked)
i=1
while i<=10:
  print('number is')
  i+=1

  if i==3:
    pass
  print(i)
```
