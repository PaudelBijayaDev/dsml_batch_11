```python
#Finding the grade of see student

cgpa=float(input("Enter your CGPA:"))

if cgpa>=0 and cgpa<=4.0:
  if cgpa<1.6:
    print("You have failed")
  elif cgpa>=1.6 and cgpa<2.0:
    print("Your grade is D")
  elif cgpa>=2.0 and cgpa<=2.4:
    print("Your grade is C")
  elif cgpa>=2.4 and cgpa<=2.8:
    print("Your grade is C+")
  elif cgpa>=2.8 and cgpa<=3.2:
    print("Your grade is B")
  elif cgpa>=3.2 and cgpa<=3.6:
    print("Your grade is B+")
  elif cgpa>=3.6 and cgpa<=4.0:
    print("Your grade is A")
  else:
    print("Your grade is A+")
else:
  print("Enter a valid CGPA")
```

# Any works as or operator

```python
#any function with if-works as or operator

#login permission
is_admin=input("are you admin.  yes/no")
is_teacher=input("are you teacher.  yes/no")

#is_admin=False
#is_teacher=True

if any([is_admin, is_teacher]):
  print('access granted')
else:
  print('access denied')
```

# All works as and operator

```python
#all function with if-works as and operator

#form validation
name=input('Enter your name')
email=input('Enter your email')
phone=input('Enter your phone number')



if all([name, email, phone]):
  print('form is valid')
else:
  print('form is invalid')
```
