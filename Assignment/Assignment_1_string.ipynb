```python
#taking sentence from user and performing different string operations on it

#1. Getting a sentence from the user
text=input("Enter a sentence: ")

#2 converting case(upper and lower)
upper_text = text.upper()
lower_text = text.lower()

#3 counting words(split into list and count the items)
words_list = text.split()
word_count = len(words_list)

#4 checking digits or alphabet
is_alpha=text.replace(" ","").isalpha()
has_digits=any(char.isdigit() for char in text)

#5 replace word-replace first word with python
if word_count > 0:
    replaced_text = text.replace(words_list[0], "Python")
else:
    replaced_text = "No words to replace"


#printing all results
print("--- String Operation Results ---")
print("Original Sentence:", text)
print("In Uppercase:     ", upper_text)
print("In Lowercase:     ", lower_text)
print("Total Word Count: ", word_count)
print("Is only letters?  ", is_alpha)
print("Contains digits?  ", has_digits)
print("Replaced version: ", replaced_text)
```
